```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Top 100 Healthiest Food in the World.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Food Origin
# Dataset
top_origin = df['Originated From'].value_counts().head(7)

# Pie Chart
plt.figure(figsize=(8,8))
top_origin.plot.pie(autopct="%1.1f%%", startangle=90)
plt.title("Most Food Origin")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Top 100 Healthiest Food in the World.csv")
df

# Others Asset
def assign_calories(calories):
    if calories <= 200:
        return "Low"
    elif 201 <= calories <= 400:
        return "Medium"
    else:
        return "High"

# Implement Calorie Level Based On Calorie
df['Calorie Level'] = df['Calories'].apply(assign_calories)

# Exploratory Data Analysis (EDA) - Share of Calorie Level for Healthiest Food
# Dataset
top_origin = df['Calorie Level'].value_counts().head(7)

# Pie Chart
plt.figure(figsize=(8,8))
top_origin.plot.pie(autopct="%1.1f%%", startangle=90)
plt.title("Share of Calorie Level for Healthiest Food")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Top 100 Healthiest Food in the World.csv")
df

# Others Asset
def assign_protein(protein):
    if protein < 10:
        return "Low"
    elif protein < 20:
        return "Medium"
    else:
        return "High"

# Implement Protein Level Based On Protein
df['Protein Level'] = df['Protein (g)'].apply(assign_protein)

# Exploratory Data Analysis (EDA) - Share of Protein Level for Healthiest Food
# Dataset
top_origin = df['Protein Level'].value_counts().head(7)

# Pie Chart
plt.figure(figsize=(8,8))
top_origin.plot.pie(autopct="%1.1f%%", startangle=90)
plt.title("Share of Protein Level for Healthiest Food")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Top 100 Healthiest Food in the World.csv")
df

# Others Asset
def assign_vit_c(vit_c):
    if vit_c < 30:
        return "Low"
    elif vit_c <= 70:
        return "Medium"
    else:
        return "High"

# Implement Vitamin C Level Based On Vitamin C
df['Vitamin C Level'] = df['Vitamin C (mg)'].apply(assign_vit_c)

# Exploratory Data Analysis (EDA) - Share of Vitamin C Level for Healthiest Food
# Dataset
top_origin = df['Vitamin C Level'].value_counts().head(7)

# Pie Chart
plt.figure(figsize=(8,8))
top_origin.plot.pie(autopct="%1.1f%%", startangle=90)
plt.title("Share of Vitamin C Level for Healthiest Food")
plt.ylabel("")
plt.show()
```
