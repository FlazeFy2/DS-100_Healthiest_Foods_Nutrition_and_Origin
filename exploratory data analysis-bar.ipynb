```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("Top 100 Healthiest Food in the World.csv")

# Exploratory Data Analysis (EDA) - Bar Chart for 7 Most Food by Context
target_col = ['Calories','Protein (g)','Fiber (g)','Vitamin C (mg)','Antioxidant Score']

for dt in target_col:
    # Dataset
    top_7_foods = df.nlargest(7, dt)[['Food', dt]]
    
    # Bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(top_7_foods['Food'], top_7_foods[dt])
    plt.title(f'Top 7 Foods by {dt}')
    plt.xlabel('Food')
    plt.ylabel(dt)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
